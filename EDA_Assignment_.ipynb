#Question 2: Explain overfitting and underfitting in ML. How can you detect and prevent them?

Hint: Discuss bias-variance tradeoff, cross-validation, and regularization techniques.

 Answer ->   Answer: In Machine Learning, the main goal of a model is to learn useful patterns from the training data and perform well on unseen (new) data. When the model becomes too complex or too simple, its performance drops. These two situations are called overfitting and underfitting.

**a. Overfitting**

Overfitting happens when a model learns too much from the training data, including noise, random fluctuations, and unnecessary details. As a result, the model performs extremely well on training data but poorly on new data.

**Characteristics of Overfitting**

Very high training accuracy

Low validation/test accuracy

Model memorizes patterns instead of generalizing

**Causes of Overfitting**

Model is too complex (deep neural networks, deep decision trees)

Too many parameters

Not enough training data

Training for too many epochs

Lack of regularization

**How to Detect Overfitting**

 1.  large gap between training and validation accuracy (Training â†‘, Validation â†“)

2. Validation loss starts increasing while training loss keeps decreasing

Learning curves diverge

**How to Prevent Overfitting**
1. Cross-Validation (e.g., k-fold CV) Helps evaluate model performance on multiple data splits.

2. Regularization Techniques

L1 (Lasso)

L2 (Ridge)

Dropout (for neural networks)

3. Early Stopping Stop training when validation loss starts rising.

4. Reduce Model Complexity Use smaller trees, fewer layers, or lower polynomial degree.

5. Data Augmentation Especially in image/text data.

6. Increase Training Data More data helps the model generalize better.

**2. Underfitting**

Underfitting happens when the model is too simple to learn the underlying patterns in the data. It performs poorly on both training and test data.

Characteristics of Underfitting
**bold text**
Low training accuracy

Low validation accuracy

Model fails to learn important patterns

**Causes of Underfitting**

Model is too simple (e.g., shallow decision tree, low-degree polynomial)

Too much regularization

Not enough training time

Important features missing

**How to Detect Underfitting**

Training accuracy is low

Validation accuracy is also low

Learning curves show both losses are high

**How to Prevent Underfitting**
1. Increase Model Complexity Deeper trees, more layers, higher polynomial degree.

2. Reduce Regularization Lower L1/L2 penalty.

3. Train Longer More epochs or more iterations.

4. Add Better Features Feature engineering or using more relevant variables.

**3. Biasâ€“Variance Tradeoff**

Underfitting = High Bias + Low Variance (Model too simple)

Overfitting = Low Bias + High Variance (Model too complex)

A good model balances both bias and variance to achieve optimal performance.

**Conclusion**

Overfitting and underfitting are two opposite problems in ML. A good model should neither memorize the data nor ignore important patterns. Techniques like cross-validation, regularization, early stopping, and proper model selection help achieve this balance.

 Answer:Handling missing values is a key step in data preprocessing because missing data can reduce model accuracy and lead to biased predictions. There are multiple methods to handle missing values depending on the data type and the amount of missing data.

**1. Deletion Methods**
a) Listwise Deletion (Removing Rows)

Rows containing missing values are removed.

Example: If only 3 out of 1000 rows have missing values, deleting them will not affect the dataset much.

When to use:

Missing data is very small (<5%)

Dataset is large

**b) Column Deletion**

Entire columns with too many missing values can be removed.

Example: If a column has 70% missing values and is not

a) Mean / Median / Mode Imputation

Replace missing values with statistical measures.

Mean â†’ for numerical data (e.g., age)

Median â†’ for skewed numerical data

Mode â†’ for categorical data

Example: If the â€œAgeâ€ column has missing values, replace them with the mean age of all available entries.

**b) Forward Fill / Backward Fill**

Used mostly in time-series data.

Forward Fill: Fill missing value with the previous value

Backward Fill: Fill missing value with the next value

Example: Stock prices dataset often uses forward fill to maintain continuity.

**3. Advanced Techniques**

**a) KNN Imputation**

Uses the values of similar (nearest) data points to fill the missing values.

Example: If height is missing, KNN finds the nearest neighbors with similar age/weight and estimates height.

**b) MICE (Multiple Imputation by Chained Equations)**

Creates multiple models to predict missing values using other features.

Example: Missing â€œsalaryâ€ can be predicted using â€œageâ€, â€œeducationâ€, â€œexperienceâ€, etc.

**Conclusion**

Handling missing values is important for accurate ML model performance. Simple methods like mean/median imputation are easy to apply, while advanced methods like KNN or MICE give better accuracy when used correctly.

Q3. What is imbalanced data in ML? Provide two techniques to handle imbalanced datasets.

**Answer:**

In Machine Learning, imbalanced data refers to a dataset where the number of samples in each class is not equal. One class has many more samples than the other(s). This causes the ML model to become biased toward the majority class, because it learns more from that class.

**Example of Imbalanced Data**

Class Number of Samples

Class 0 (Majority) 950 Class 1 (Minority) 50

Here, Class 0 dominates the dataset. A normal classifier may predict everything as Class 0 and still show 95% accuracy, but it completely fails to detect minority cases.

T**his is a common issue in**:

Fraud detection

Medical diagnosis

Spam detection

Churn prediction

Why Imbalanced Data is a Problem?

Model becomes biased toward majority class
Poor recall for minority class

Accuracy becomes misleading

Rare but important cases are missed (fraud, disease, etc.)

Techniques to Handle Imbalanced Data

Below are two commonly used and effective techniques:

1. SMOTE (Synthetic Minority Oversampling Technique)
SMOTE increases the size of the minority class by creating synthetic samples, not just duplicating existing ones.


H**ow SMOTE Works**

It selects a minority class sample

Finds its nearest neighbors

Creates new synthetic data points between them

**Advantages**

Prevents overfitting compared to simple oversampling

Works well with continuous data

2.Random Undersampling

This method reduces the majority class size by randomly removing samples.

How Undersampling Works

Selects majority class

Randomly deletes some data points

Balances both classes

Advantages

Fast and simple

Reduces memory usage

Works well when dataset is large

Disadvantages

May remove important samples

Can reduce model accuracy

Other Techniques
.
3.Class Weighting

Assign higher weight to the minority class during model training. Used in:

Logistic Regression

4. Random Forest

SVM

Neural Networks

Ensemble Methods
Balanced Random Forest

EasyEnsemble These combine multiple models to reduce bias.

5. Cost-Sensitive Learning

Penalizes the model more for misclassifying the minority class.

**Conclusion**

Imbalanced datasets can seriously affect the performance of ML models. Techniques such as SMOTE and Random Undersampling help balance the data and improve the modelâ€™s ability to correctly classify minority cases.

#Question 6: Compare Label Encoding and One-Hot Encoding. When would you prefer
one over the other?
Hint: Consider categorical variables with ordinal vs. nominal relationships.

 Answer ->
 **Explanation:**
Label Encoding converts categorical values into numerical labels.
Each unique category is assigned a number (0, 1, 2, â€¦).


**Example:**
Low â†’ 0

Medium â†’ 1

High â†’ 2

**When to Use**:

Label Encoding is preferred when the categorical variable has an ordinal relationship,
where the order of categories matters.

**Advantages:**

Simple and fast

Uses less memory

Disadvantages:

Can mislead models if no natural order exists

One-Hot Encoding

Explanation:
**bold text**
One-Hot Encoding converts each category into a separate binary column.
Each column contains 0 or 1 to represent the presence of a category.

Example:

Color â†’ Red, Blue, Green

Red	Blue	Green
1	0	0
0	1	0

When to Use:

One-Hot Encoding is preferred when the categorical variable has a nominal relationship,
where no order exists between categories.

**Advantages:**
Prevents false ordering

Works well with most ML algorithms

Disadvantages:

Increases number of features

Can cause high dimensionality

Comparison Table

Feature	Label Encoding	One-Hot Encoding
Output	Single numeric column	Multiple binary columns
Order Assumption	Yes	No
Best for	Ordinal data	Nominal data
Feature Size	Small	Large

**Conclusion**

Label Encoding should be used for ordinal categorical variables where order is important.
One-Hot Encoding should be used for nominal categorical variables where no order exists.

#Question 7: Google Play Store Dataset
a). Analyze the relationship between app categories and ratings. Which categories have the
highest/lowest average ratings, and what could be the possible reasons?
Dataset: https://github.com/MasteriNeuron/datasets.git
(Include your Python code and output in the code box below.)


Answer ->  

**Objective**

The goal of this analysis is to understand the relationship between app categories and user ratings in the Google Play Store dataset.
We calculate the average rating for each app category and identify which categories have the highest and lowest ratings.



```python
import pandas as pd

# Load dataset
df = pd.read_csv("googleplaystore.csv")

# Clean Rating column
df['Rating'] = pd.to_numeric(df['Rating'], errors='coerce')
df = df.dropna(subset=['Rating'])

# Group and analyze
category_ratings = df.groupby('Category')['Rating'].mean().sort_values(ascending=False)

print("Top 5 Categories with Highest Average Ratings:")
print(category_ratings.head())

print("\nBottom 5 Categories with Lowest Average Ratings:")
print(category_ratings.tail())
```

Analysis and Interpretation

Users have clear expectations

**Categories with Lowest Average Ratings**

Dating

Tools

**Possible Reasons:**

High competition in these categories

Too many ads and in-app purchases

Performance issues and bugs

User dissatisfaction

**Conclusion**

The analysis shows that educational and informational app categories generally receive higher ratings, while games and utility-based apps tend to have lower average ratings.

#Question 8: Titanic Dataset
a) Compare the survival rates based on passenger class (Pclass). Which class had the highest
survival rate, and why do you think that happened?
b) Analyze how age (Age) affected survival. Group passengers into children (Age < 18) and
adults (Age â‰¥ 18). Did children have a better chance of survival?
Dataset: https://github.com/MasteriNeuron/datasets.git
(Include your Python code and output in the code box below.)


  Answer ->
  Objective

The objective of this analysis is:

(a) To compare survival rates based on passenger class (Pclass)


(b) To analyze how age affected survival, comparing children vs adults


```python
import pandas as pd

# Load Titanic dataset from GitHub
url = "https://raw.githubusercontent.com/MasteriNeuron/datasets/master/titanic.csv"
df = pd.read_csv(url)

# -----------------------------
# Part (a): Survival by Pclass
# -----------------------------

# Calculate survival rate for each passenger class
pclass_survival = df.groupby('Pclass')['Survived'].mean()

print("Survival Rate by Passenger Class:")
print(pclass_survival)

# -----------------------------
# Part (b): Survival by Age Group
# -----------------------------

# Handle missing Age values
df['Age'] = df['Age'].fillna(df['Age'].median())

# Create Age Group column
df['AgeGroup'] = df['Age'].apply(lambda x: 'Child' if x < 18 else 'Adult')

# Calculate survival rate by age group
age_survival = df.groupby('AgeGroup')['Survived'].mean()

print("\nSurvival Rate by Age Group:")
print(age_survival)
```

   a) Survival Rate Based on Passenger Class (Pclass)
Observation

1st Class passengers had the highest survival rate

3rd Class passengers had the lowest survival rate

**Reason**

First-class passengers had better access to lifeboats

They were located closer to decks

Priority was given to higher-class passengers during evacuation


(b) Survival Rate Based on Age (Children vs Adults)
Observation

Children had a higher survival rate than adults

Reason

The evacuation policy followed â€œWomen and children firstâ€

Children were prioritized for lifeboats

Adults, especially men, had lower chances of survival


Conclusion
**bold text**
Passenger class played a major role in survival, with 1st class passengers surviving the most

Children had a better chance of survival than adults

#Question 9: Flight Price Prediction Dataset
a) How do flight prices vary with the days left until departure? Identify any exponential price
surges and recommend the best booking window.
b)Compare prices across airlines for the same route (e.g., Delhi-Mumbai). Which airlines are
consistently cheaper/premium, and why?
Dataset: https://github.com/MasteriNeuron/datasets.git
(Include your Python code and output in the code box below.)


Answer ->  
Objective

(a) To analyze how flight prices change with days left before departure

(b) To compare prices across different airlines for the same route (Delhiâ€“Mumbai)


```python
import pandas as pd

# Load Flight Price dataset from GitHub
url = "https://raw.githubusercontent.com/MasteriNeuron/datasets/master/flight_price.csv"
df = pd.read_csv(url)

# ---------------------------------------------
# Part (a): Price vs Days Left Until Departure
# ---------------------------------------------

# Calculate average price by days left
days_price = df.groupby('days_left')['price'].mean().sort_index()

print("Average Flight Price by Days Left Until Departure:")
print(days_price.head(10))
print("\nLast 10 days before departure:")
print(days_price.tail(10))

# ---------------------------------------------
# Part (b): Airline Price Comparison (Delhi-Mumbai)
# ---------------------------------------------

# Filter Delhi to Mumbai route
route_df = df[(df['source_city'] == 'Delhi') & (df['destination_city'] == 'Mumbai')]

# Average price by airline
airline_prices = route_df.groupby('airline')['price'].mean().sort_values()

print("\nAverage Flight Price by Airline (Delhi-Mumbai Route):")
print(airline_prices)
```

(a) How Flight Prices Vary with Days Left

Prices are stable and comparatively cheaper

**Reason**

Airlines increase prices as seats fill up

Last-minute travelers are willing to pay more


b) Price Comparison Across Airlines (Delhiâ€“Mumbai)
Observation

Low-cost airlines (AirAsia, IndiGo, SpiceJet) are cheaper

Premium airlines (Vistara, Air India) are more expensive

#Question 10: HR Analytics Dataset
a). What factors most strongly correlate with employee attrition? Use visualizations to show key
drivers (e.g., satisfaction, overtime, salary).
b). Are employees with more projects more likely to leave?
Dataset: hr_analytics


 Answer ->   Objective

a) To identify the key factors that strongly influence employee attrition

b) To analyze whether employees with more projects are more likely to leave the company


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
url = "https://raw.githubusercontent.com/MasteriNeuron/datasets/master/hr_analytics.csv"
df = pd.read_csv(url)

# -------------------------------
# (a) Factors affecting Attrition
# -------------------------------

# Attrition vs Satisfaction Level
satisfaction_attrition = df.groupby('satisfaction_level')['left'].mean()

satisfaction_attrition.plot(kind='line')
plt.title("Attrition Rate vs Satisfaction Level")
plt.xlabel("Satisfaction Level")
plt.ylabel("Attrition Rate")
plt.show()

# Attrition vs Salary
salary_attrition = df.groupby('salary')['left'].mean()

salary_attrition.plot(kind='bar')
plt.title("Attrition Rate by Salary Level")
plt.xlabel("Salary Level")
plt.ylabel("Attrition Rate")
plt.show()

# -------------------------------
# (b) Projects vs Attrition
# -------------------------------

projects_attrition = df.groupby('number_project')['left'].mean()

projects_attrition.plot(kind='line')
plt.title("Attrition vs Number of Projects")
plt.xlabel("Number of Projects")
plt.ylabel("Attrition Rate")
plt.show()

print("Attrition Rate by Number of Projects:")
print(projects_attrition)
```

Output (Sample Interpretation)

Employees with low satisfaction levels show a much higher attrition rate

Employees with low salary leave more compared to medium and high salary employees

Attrition increases sharply when employees work on very high number of projects (6â€“7 projects)
